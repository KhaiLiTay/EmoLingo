```python
import librosa
def zcr(data,frame_length,hop_length):
    zcr=librosa.feature.zero_crossing_rate(data,frame_length=frame_length,hop_length=hop_length)
    return np.squeeze(zcr)
def rmse(data,frame_length=2048,hop_length=512):
    rmse=librosa.feature.rms(y=data,frame_length=frame_length,hop_length=hop_length)
    return np.squeeze(rmse)
def mfcc(data,sr,frame_length=2048,hop_length=512,flatten:bool=True):
    mfcc=librosa.feature.mfcc(y=data,sr=sr)
    return np.squeeze(mfcc.T)if not flatten else np.ravel(mfcc.T)

def extract_features(data,sr=22050,frame_length=2048,hop_length=512):
    result=np.array([])

    result=np.hstack((result,
                      zcr(data,frame_length,hop_length),
                      rmse(data,frame_length,hop_length),
                      mfcc(data,sr,frame_length,hop_length)
                     ))
    return result
```

```python
def get_predict_feat(path):
    d, s_rate= librosa.load(path, duration=2.5, offset=0.6)
    res=extract_features(d)
    result=np.array(res)
    result=np.reshape(result,newshape=(1,2376))
    i_result = scaler2.transform(result)
    final_result=np.expand_dims(i_result, axis=2)

    return final_result
```

# STT

```python
def speech_to_text(audio_file, speaker_language):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio_data = recognizer.record(source)
    try:
        text = recognizer.recognize_google(audio_data, language=speaker_language)
        return text
    except sr.UnknownValueError:
        return "Sorry, I couldn't understand the audio."
    except sr.RequestError:
        return "Sorry, there was a request error."
```

```python
def gpt_35_api_stream(messages: list, language: str = "en", emotion: str = None):
    # Set system messages based on incoming language parameters
    language_context = {
        "en": f"You are an assistant that replies in English. Please respond with empathy considering the user's emotion.",
        "zh-cn": f"你是一个使用中文回复的助手。请根据用户的情绪提供有同理心的回复。",
        "es": f"Eres un asistente que responde en español. Responde con empatía considerando la emoción del usuario.",
        "fr": f"Vous êtes un assistant qui répond en français. Répondez avec empathie en tenant compte de l'émotion de l'utilisateur.",
        "de": f"Du bist ein Assistent, der auf Deutsch antwortet. Bitte antworte einfühlsam und berücksichtige die Emotion des Benutzers.",
        "ja": f"日本語で返信するアシスタントです。ユーザーの感情を考慮して共感的に返信してください。",
        "ko": f"한국어로 응답하는 어시스턴트입니다. 사용자 감정을 고려하여 공감으로 응답하십시오。",
        "hi": f"आप एक सहायक हैं जो हिंदी में जवाब देता है। कृपया उपयोगकर्ता की भावना को ध्यान में रखते हुए सहानुभूति के साथ जवाब दें।"
    }
    system_message = language_context.get(language, "You are an assistant that replies in English. Please respond with empathy.")

    # Inject emotion into context and enhance logic
    if emotion:
        emotion_guidelines = {
            "Neutral": "Respond normally based on the user's content.",
            "Calm": "Respond in a composed and supportive manner.",
            "Happy": "Share their joy and respond with an uplifting and enthusiastic tone.",
            "Sad": "Provide comforting and encouraging words to uplift their mood.",
            "Angry": "Help them calm down and encourage a rational perspective. Avoid escalating the situation.",
            "Fearful": "Offer reassurance and words of encouragement to help them overcome their fears.",
            "Disgust": "Address their concerns respectfully and provide supportive and understanding feedback.",
            "Surprised": "Encourage them to share more details and join in their excitement."
        }
        # Adjust system messages based on emotion type
        emotion_instruction = emotion_guidelines.get(emotion, "Respond with empathy considering the user's emotion.")
        system_message += f" The user's detected emotion is '{emotion}'. {emotion_instruction}"

    # Generate canned responses directly to unintelligible audio
    if messages and "Sorry, I couldn't understand the audio." in messages[0]["content"]:
        predefined_replies = {
            "en": "Sorry, I couldn't understand what you said. Could you repeat it one more time?",
            "zh-cn": "抱歉，我不理解你在说什么，可以重复多一次吗？",
            "es": "Lo siento, no pude entender lo que dijiste. ¿Podrías repetirlo otra vez?",
            "fr": "Désolé, je n'ai pas pu comprendre ce que vous avez dit. Pourriez-vous répéter encore une fois ?",
            "de": "Entschuldigung, ich konnte nicht verstehen, was du gesagt hast. Könntest du es bitte noch einmal wiederholen?",
            "ja": "申し訳ありませんが、何を言っているのかわかりませんでした。もう一度繰り返していただけますか？",
            "ko": "죄송합니다. 무슨 말을 하시는지 이해하지 못했습니다. 다시 한번 말씀해 주시겠습니까?",
            "hi": "क्षमा करें, मैं समझ नहीं पाया कि आपने क्या कहा। क्या आप इसे फिर से दोहरा सकते हैं?"
        }
        return predefined_replies.get(language, "Sorry, I couldn't understand what you said. Could you repeat it one more time?")

    # Insert system message
    messages.insert(0, {"role": "system", "content": system_message})

    # Call GPT API (Streaming)
    reply = ""
    stream = client.chat.completions.create(
        model='gpt-3.5-turbo',
        messages=messages,
        stream=True,
    )
    for chunk in stream:
        if chunk.choices[0].delta.content is not None:
            content = chunk.choices[0].delta.content
            reply += content
    return reply
```

# TTS

```python
def generate_tts_audio(text: str, emotion: str, language: str, speaker_wav_path: str):

    supported_languages = ['en', 'es', 'fr', 'de', 'it', 'pt', 'pl', 'tr',
                           'ru', 'nl', 'cs', 'ar', 'zh-cn', 'hu', 'ko', 'ja', 'hi']


    # Set output directory
    output_dir_colab = "./output/"
    output_dir_drive = "/content/drive/MyDrive/ML Team 35/Codes/Chatbot/Generated_Audio/"
    os.makedirs(output_dir_colab, exist_ok=True)
    os.makedirs(output_dir_drive, exist_ok=True)

    # Set file path
    output_file_path_colab = os.path.join(output_dir_colab, f"output_{emotion}.wav")
    output_file_path_drive = os.path.join(output_dir_drive, f"output_{emotion}.wav")

    # If the language is not supported, generate default English audio
    if language not in supported_languages:
        print(f"Language '{language}' is not supported. Falling back to default English audio.")
        text = "Sorry, I couldn't understand your language. Could you try speaking in another language?"
        language = "en"
        emotion = "Neutral"

    try:
        # generate audio
        start = time.perf_counter()
        tts.tts_to_file(
            text=text,
            speaker_wav=speaker_wav_path,
            language=language,
            file_path=output_file_path_colab,
            emotion=emotion,
        )
        end = time.perf_counter()

        # Copy files to Google Drive
        shutil.copy(output_file_path_colab, output_file_path_drive)

        # output log messages
        print(f"Audio file saved to Colab: {output_file_path_colab}")
        print(f"Audio file also saved to Drive: {output_file_path_drive}")
        print(f"Generation time: {end - start:.2f} seconds")

        return output_file_path_colab, output_file_path_drive
    except AssertionError as e:
        print(f"Error generating audio: {e}")
        return None, None
```

# Chatbot


```python
def chatbot_voice_interaction(record_duration=5):
    """
    Complete voice chatbot process: recording -> STT -> SER -> GPT reply -> TTS -> play the generated voice

    Args: record_duration (int): recording duration (seconds).

    Returns: str: Returns the reply text generated by GPT.
    """
    # Step 1: Record and save the audio file

    print("Recording audio...")
    # with tqdm(total=record_duration, desc="Recording Progress", unit="s") as pbar:
    #     for _ in range(record_duration):
    #         time.sleep(1)
    #         pbar.update(1)
    # record audio
    audio = record(sec=record_duration)

    print("Recording completed.")

    audio_path_colab = "/content/recorded_audio.wav"
    audio_path_drive = "/content/drive/MyDrive/ML Team 35/Codes/Chatbot/recorded_audio.wav"
    audio.export(audio_path_colab, format="wav")
    audio.export(audio_path_drive, format="wav")
    print(f"Audio saved to Colab: {audio_path_colab}")
    print(f"Audio saved to Google Drive: {audio_path_drive}")
    display(Audio(audio_path_colab))

    # Step 2: Detect language
    print("Transcribing and detecting language...")
    result = model.transcribe(audio_path_colab, task="language")
    speaker_language = result['language']
    if speaker_language == "zh":
        speaker_language = "zh-cn"
    print(f"Detected language: {speaker_language}")

    # Step 3: STT (speech to text)
    print("Converting speech to text...")
    stt_result = speech_to_text(audio_path_colab, speaker_language)
    print(f"STT Result: {stt_result}")

    # Step 4: SER (emotion recognition)
    print("Detecting speech emotion...")
    speech_emotion = get_ser_label(audio_path_colab)
    print(f"Speech Emotion: {speech_emotion}")

    # Step 5: Call the GPT streaming API to generate a reply
    print("Generating GPT reply...")
    summarizing_factor = "(reply in one sentence with more concerning with speaker's emotion)"
    messages = [{'role': 'user', 'content': stt_result + summarizing_factor + speech_emotion}]
    gpt_reply = gpt_35_api_stream(messages, language=speaker_language, emotion=speech_emotion)
    print(f"GPT Reply: {gpt_reply}")

    # Step 6: TTS (Text to Speech)
    print("Generating audio response...")
    speaker_wav = "/content/drive/MyDrive/ML Team 35/Codes/Chatbot/LiShen.wav"
    output_file_path_colab, output_file_path_drive = generate_tts_audio(
        text=gpt_reply,
        emotion=speech_emotion,
        language=speaker_language,
        speaker_wav_path=speaker_wav
    )

    # Step 7: Play the generated speech
    if os.path.exists(output_file_path_colab):
        print(f"Playing generated audio file: {output_file_path_colab}")
        display(Audio(output_file_path_colab))
    else:
        print("Generated audio file not found. Please check the paths and settings.")

    return gpt_reply
```
